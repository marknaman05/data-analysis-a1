```python
column_data_types = {'Latitude': str, 'Longitude': str}
data = pd.read_csv('/content/pincode.csv', dtype=column_data_types)

print(data.dtypes)
```

```python
data_cleaned = data.dropna()
data_cleaned
```

```python
data_cleaned.reset_index(drop=True, inplace=True)
data_cleaned
```

```python
while True:
    try:
        data['Latitude'] = data['Latitude'].astype(float)
        data['Longitude'] = data['Longitude'].astype(float)
        print("Conversion succeeded")
        break  # If conversion succeeds, exit the loop
    except ValueError as e:
        print(f"Error: {e}")
        print("Cleaning and retrying the conversion...")
        data['Latitude'] = data['Latitude'].str.extract('(-?\d+\.\d+)').astype(float)
        data['Longitude'] = data['Longitude'].str.extract('(-?\d+\.\d+)').astype(float)
        continue
```

```python
# types of each column of the table
print(data.dtypes)
```
