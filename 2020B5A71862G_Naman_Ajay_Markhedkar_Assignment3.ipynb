# Data Visualization

```python
melted_df = pd.melt(transposed_df, id_vars=['Date'], value_vars=['Temperature Min', 'Temperature Max'],
                    var_name='Temperature Type', value_name='Temperature (°C)')

melted_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates


plt.figure(figsize=(12, 6))
sns.lineplot(data=melted_df, x='Date', y='Temperature (°C)', hue='Temperature Type')
plt.title('Min Max Temperature in Kapra Mandal from July 2013 to Januray 2018')
plt.xlabel('Months')
plt.ylabel('Temperature in °C')


date_format = mdates.DateFormatter("%b %Y")
plt.gca().xaxis.set_major_formatter(date_format)

plt.grid(True)

plt.show()
```
