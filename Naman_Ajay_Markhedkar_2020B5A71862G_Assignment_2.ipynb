```python
import pandas as pd

file_path = '/content/200 participant collected data.xlsx'

# Read the Excel file
df=pd.read_excel("200 participant collected data.xlsx", usecols="B,E")
df
```

# Data Cleaning

```python
gender_counts = df['Gender'].value_counts()

# Access the count for 'Male' and 'Female'
male_count = gender_counts.get('Male', 0)
female_count = gender_counts.get('Female', 0)

print(f"Number of Males: {male_count}")
print(f"Number of Females: {female_count}")
```

```python
df['Gender']=df['Gender'].str.capitalize()
```

```python
# Count no of Males in Gender column

gender_counts = df['Gender'].value_counts()

# Access the count for 'Male' and 'Female'
male_count = gender_counts.get('Male', 0)
female_count = gender_counts.get('Female', 0)

print(f"Number of Males: {male_count}")
print(f"Number of Females: {female_count}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a figure with subplots
fig, axes = plt.subplots(3, 1, figsize=(8, 15))

# Histogram plot
sns.histplot(data=df, x='Height (cm) (average of 3 measurments)', hue='Gender', kde=True, ax=axes[0])
axes[0].set_title('Histogram')

# Box plot
sns.boxplot(data=df, x='Height (cm) (average of 3 measurments)', y='Gender', ax=axes[1])
axes[1].set_title('Box Plot')

# Violin plot
sns.violinplot(data=df, x='Height (cm) (average of 3 measurments)', y='Gender', ax=axes[2])
axes[2].set_title('Violin Plot')

plt.tight_layout()
plt.show()
```
